# Tensors:
* In mathematics and computer science, a tensor is a multi-dimensional array that generalizes the concepts of scalars, vectors, and matrices. Tensors are used extensively in fields such as physics, engineering, machine learning, and more
* pytorch is library for processing tensors.
* a tensor is number, vector, matrix or any n-diamensional array.
* Tensor is homogenous datatype check vector data
* each row & columns has equal no. of items

# 5. Random Tensor- 
* it is nothing but tensor with random values in range 0-1
* we need to also specify the size of tensor.


```python
# creating Tensor with random values in range of 0-1
random_tensor = t.rand(size= (3,4))  # 3-rows & 4-columns
```

```python
random_tensor
```

```python
# bigger Tensor by using random values
bigger_tensor = t.rand(3, 4, 5)  

bigger_tensor
```

### 8. Creating tensor_like with same diamension but zeros & onces values

```python
sample = t.rand(1, 3, 4)
sample
```

```python
# creating zeros tensor with same dimension of sample 
zero_like = t.zeros_like(sample)
zero_like
```

```python
# creating ones tensor with same dimension of sample 
ones_like = t.ones_like(sample)
ones_like
```
